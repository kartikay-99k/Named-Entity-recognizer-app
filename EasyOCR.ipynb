```python
def img_to_text(img):
    bounds = reader.readtext(img)
    final_text = ' '.join([str(i[1]) for i in bounds])
    return final_text
```

```python
disallowed_characters = "|_”!‘@€`~é(«]»[#}“{)"""

import spacy
nlp=spacy.load("en_core_web_sm")

def preprocess(tx):
    for character in disallowed_characters:
        tx = tx.replace(character, "")
    tx= tx.replace('\n', " ")
    tx= tx.replace("—", " ")
    my_doc = nlp(tx)
    my_doc_cleaned = [token for token in my_doc if not token.is_stop]
    listToStr = ' '.join([str(elem) for elem in my_doc_cleaned])
    return listToStr
```

```python
tx = img_to_text("33.jpg")
tx= preprocess(tx)
tx
```
